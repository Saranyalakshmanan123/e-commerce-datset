```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("ecommerce Dataset.xlsx.csv")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.figure(figsize=(10, 6))
plt.scatter(df['Amount'], df['Profit'], alpha=0.6)
plt.title('Amount vs Profit (using Matplotlib)')
plt.xlabel('Amount')
plt.ylabel('Profit')
plt.show()
```

```python
print("Descriptive statistics for numerical columns:")
display(df[['Amount', 'Profit', 'Quantity']].describe())

print("\nNumber of unique values for categorical columns:")
print("Category:", df['Category'].nunique())
print("Sub-Category:", df['Sub-Category'].nunique())
print("PaymentMode:", df['PaymentMode'].nunique())

print("\nFrequency distribution for categorical columns:")
print("\nCategory:")
display(df['Category'].value_counts())
print("\nSub-Category:")
display(df['Sub-Category'].value_counts())
print("\nPaymentMode:")
display(df['PaymentMode'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
sns.set_style("whitegrid")

# 1. Histograms for numerical columns
numerical_cols = ['Amount', 'Profit', 'Quantity']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, col in enumerate(numerical_cols):
    sns.histplot(data=df, x=col, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 2. Bar plots for categorical columns
categorical_cols = ['Category', 'Sub-Category', 'PaymentMode']
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
for i, col in enumerate(categorical_cols):
    sns.countplot(data=df, x=col, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Frequency of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    if col == 'Sub-Category':
        axes[i].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 4. Scatter plot of 'Amount' vs 'Profit' colored by 'Category'
plt.figure(figsize=(12, 7))
sns.scatterplot(data=df, x='Amount', y='Profit', hue='Category', alpha=0.6)
plt.title('Amount vs Profit colored by Category')
plt.xlabel('Amount')
plt.ylabel('Profit')
plt.legend(title='Category')
plt.show()
```
